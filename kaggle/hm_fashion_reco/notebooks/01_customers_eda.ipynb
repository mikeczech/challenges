# Explore Customer Data

```python
df_customers = pd.read_csv("../data/customers.csv")
```

```python
df_customers
```

```python
df_customers["customer_id"].nunique()
```

```python
df_customers.isna().mean().plot.bar(title="NA Ratios")
```

```python
categorical_cols = ["FN", "Active", "club_member_status", "fashion_news_frequency"]

fig, axs = plt.subplots(1, len(categorical_cols), figsize=(20, 4))
for i, c in enumerate(categorical_cols):
    df_customers[c].value_counts().plot.bar(ax=axs[i], title=c)
```
